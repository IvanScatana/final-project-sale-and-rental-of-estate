```python
import pandas as pd

pd.set_option("display.max_columns", None)
```

```python
df1 = pd.read_csv("../../tmp_data/bek_processed_part_1.csv")
df2 = pd.read_csv("../../tmp_data/vanya_processed_part_1.csv")
```

```python
merged_df = pd.merge(df1, df2, on="id", how="inner")

price_p1 = merged_df["price"].quantile(0.01)
price_p99 = merged_df["price"].quantile(0.99)
area_p1 = merged_df["total_area"].quantile(0.01)
area_p99 = merged_df["total_area"].quantile(0.99)

print(f"Цена: от {price_p1:,.0f} до {price_p99:,.0f}")
print(f"Площадь: от {area_p1:.1f} до {area_p99:.1f}")

mask = (
    (merged_df["price"] >= price_p1)
    & (merged_df["price"] <= price_p99)
    & (merged_df["total_area"] >= area_p1)
    & (merged_df["total_area"] <= area_p99)
)
merged_df = merged_df[mask]

merged_df.to_csv("../../tmp_data/merged_part_1.csv", index=False)

merged_df
```
