```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df1 = pd.read_csv("../../tmp_data/bek_processed_part_1.csv")
df2 = pd.read_csv("../../tmp_data/vanya_processed_part_1.csv")
```

```python
merged_df = pd.merge(df1, df2, on='id', how='inner')

mask = (merged_df['price'] <= 1_600_000) & (merged_df['total_area'] <= 450)
merged_df = merged_df[mask]

merged_df.to_csv('../../tmp_data/merged_part_1.csv', index=False)

merged_df.head()
```
